**Exemplos de trabalho com datassets com o Pandas**

```
#Importar bilbiotecas necessárias
import numpy as np
import pandas as pd
```

```
#O primeiro dataset que vamos carregar é o dataset com flores Iris, cuja historia pode consultar em https://en.wikipedia.org/wiki/Iris_flower_data_set
#Uma das bibliotecas que contem este dataset é a biblioteca scikit-learn, uma biblioteca para machine learning em Python
#Importar o package
from sklearn.datasets import load_iris

#Carregar o dataset
data = load_iris()
#Mostrar a descrição do dataset
print(data.DESCR)
```

```
#Mostrar as características do dataset (atributos)
print(data.feature_names)
```

```
#Mostrar a coluna que possui a classificação do data set
print(data.target_names)
```

```
print(data)
```

```
#Carregar o dataset para um pd
df1 = pd.DataFrame(data.data, columns=data.feature_names)
df1.head()
```

```
#Ter uma ideia da dimensão do dataset
df1.shape
```

```
#Resumo estatístico do DataFrame, com quartis, mediana, etc.
df1.describe()
```

```
#Média dos valores
df1.mean()
```

```
#Mediana dos valores
df1.median()
```

```
#Também pode visualizar estatísticas referentes a uma só coluna
df1['sepal length (cm)'].mean()
```

```
#Ordene a DataFrame pelo comprimento da sépala
df1.sort_values(by=['sepal length (cm)','sepal width (cm)'])
```

```
#Carregar o dataset para um pd com as variaveis target
df2 = pd.DataFrame(data= np.c_[data['data'], data['target']],
                     columns= data['feature_names'] + ['target'])
print(df2.head())
#Adicionar uma coluna com a descrição das especies/classes das Iris
df2['species'] = pd.Categorical.from_codes(data.target, data.target_names)
print(df2.head())
```

Pode utilizar as bibliotecas gráficas para fazer representações dos dados presentes num DataFrame.


```
#Visualizar a relação entre o comprimento e a largura da sépala das flores Iris presentes no DataFrame df2
import matplotlib.pyplot as plt
df3.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```
#Olhando para o gráfico acima ficamos sem perceber a relação entre comprimento e largura das sépalas
#Tem um primeira zona de pontos que parecem indicar que quanto maior é a largura maior é o comprimento, mas o resto não se percebe
#A razão está no facto de não termos separado as flores por raça
#Se criarmos três datasets um para cada tipo e mapearmos vamos ter uma ideia diferente
df4 = df3[df2['Species']=='Iris-setosa']
print(df4.head())
df4.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
df5 = df3[df3['Species']=='Iris-versicolor']
print(df5.head())
df5.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
df6 = df3[df3['Species']=='Iris-virginica']
print(df6.head())
df6.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```
#Olhando para os gráficos acima já consegue perceber que existe uma relação entre largura e comprimento das sépalas
#A biblioteca SeaBorn baseada no biblioteca matplotlib possui muitas opções gráficas para representação de dados https://seaborn.pydata.org/
import seaborn as sns
sns.FacetGrid(df3, hue="Species", height=5).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
```

```
#Olhando para os gráficos acima já consegue perceber que existe uma relação entre largura e comprimento das sépalas
#A biblioteca SeaBorn baseada no biblioteca matplotlib possui muitas opções gráficas para representação de dados https://seaborn.pydata.org/
import seaborn as sns
sns.FacetGrid(df3, hue="Species", height=5).map(plt.scatter, "PetalLengthCm", "PetalWidthCm").add_legend()
```

```
#Ex5. Determine a pessoa mais nova, mais velha e a idade média das pessoas que estavam no titanic
print("Mais novo:", df5['age'].min())
print("Mais velho:", df5['age'].max())
print("Média:", df5['age'].mean())
#Dá erro, porque os valores estão como texto
```

```
#Assim, temos de fazer duas coisas:
#1. Substituir ? por o valor np.NaN
df5['age']=df5['age'].replace('?',np.NaN)
#2. Converter para numerico
df5['age']=pd.to_numeric(df5['age'])
print("Mais novo:", df5['age'].min())
print("Mais velho:", df5['age'].max())
print("Média:", df5['age'].mean())
```

```
#Ex6. Quantas pessoas iam no barco com menos de 10 anos (incluindo)?
under10 = df5[df5['age']<=10]
under10.count()
under10.head(10)
```

```
#Ex8. Qual o valor médio da passagem?
#Antes de fazer este exercício é necessário converter o campo para numérico
df5['fare']=df5['fare'].replace('?',np.NaN)
df5['fare'] = pd.to_numeric(df5['fare'])
print("Preço médio do bilhete:", df5['fare'].mean())
```
